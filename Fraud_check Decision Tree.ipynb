# Convert Object values to Numeric values

```python
fraud.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
fraud['Undergrad'] = label_encoder.fit_transform(fraud['Undergrad'])
fraud['Marital.Status'] = label_encoder.fit_transform(fraud['Marital.Status'])
fraud['Urban'] = label_encoder.fit_transform(fraud['Urban'])
```

```python
fraud.info()
```

```python
fraud.rename(columns = {'Marital.Status' : 'marital_status', 'Taxable.Income': 'tax_in', 'City.Population': 'city_popu',
                    'Work.Experience': 'work_exp'}, inplace= True)
```

```python
fraud['status'] = fraud['tax_in'].apply(lambda tax_in: 'Risky' if tax_in <= 30000 else 'Good')
```

```python
fraud.head()
```

```python
fraud.drop(['tax_in'], axis=1, inplace=True)
```

```python
fraud['status'] = label_encoder.fit_transform(fraud['status'])
```

```python
X = fraud.iloc[:,0:5]
y = fraud.iloc[:,-1]
```

```python
X
```

```python
y
```

# Split into Training and Testing 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=0)
```

# Model Creation

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
```

# Model Validation: Confusion Matrix
Training and Testing accuracy 

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
test_acc = accuracy_score(y_test,y_pred)
test_acc
```

```python
y_pred1 = model.predict(X_train)
train_acc = accuracy_score(y_train,y_pred1)
train_acc 
```

# Plot the Tree

```python
from sklearn import tree
```

```python
plt.figure(figsize=(25,15))
tree.plot_tree(model,filled=True)

plt.show()
```

# K-fold testing

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
fold = KFold(n_splits=5, shuffle=False)
```

# Final Model

```python
final_model = DecisionTreeClassifier(criterion="entropy",max_depth=3)
```

```python
final_model.fit(X_train,y_train)
```

```python
plt.figure(figsize=(25,15))
tree.plot_tree(final_model,filled=True)

plt.show()
```
